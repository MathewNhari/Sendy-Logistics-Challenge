```python
def diagnostic_plots(df,variable):
    #function takes a dataframe (df) and the variable
    #of interest as arguments
    
    #define figure size
    plt.figure(figsize=(16,4))
    
    #histogram
    plt.subplot(1,3,1)
    sns.distplot(df[variable],bins=30)
    plt.title('Histogram')
    
    #Q-Q plot
    plt.subplot(1,3,2)
    stats.probplot(df[variable], dist='norm', plot=plt)
    
    #boxplot
    plt.subplot(1,3,3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')
    
    return plt.show()
    
    
```

```python
def seperate_time(df, variables):
    #takes a time column and extracts the hour minute and seconds then deletes
    #the original column
    
    for var in variables:
        df[var+'_hours'] = pd.to_datetime(df[var]).dt.hour
        df[var+'_minutes'] = pd.to_datetime(df[var]).dt.minute
        df[var+'_seconds'] = pd.to_datetime(df[var]).dt.second
    df.drop(variables,axis=1,inplace=True)
    return df
```

```python
# Calculate correlations between predictor variables and the response variable
corrs = train_df.corr()['Time from Pickup to Arrival'].sort_values(ascending=False)
```

```python
from scipy.stats import pearsonr

# Build a dictionary of correlation coefficients and p-values
dict_cp = {}

column_titles = [col for col in corrs.index if col!= 'Time from Pickup to Arrival']
for col in column_titles:
    p_val = round(pearsonr(train_df[col], train_df['Time from Pickup to Arrival'])[1],6)
    dict_cp[col] = {'Correlation_Coefficient':corrs[col],
                    'P_Value':p_val}

df_cp = pd.DataFrame(dict_cp).T
df_cp_sorted = df_cp.sort_values('P_Value')
df_cp_sorted[df_cp_sorted['P_Value']<0.1]
```

```python
# The dependent variable remains the same:
y_data = train_df['Time from Pickup to Arrival'] 

# Model building - Independent Variable (IV) DataFrame
X_names = list(df_cp[df_cp['P_Value'] < 0.05].index)
X_data = train_df[X_names]
```

```python
corr = X_data.corr()
# Find rows and columns where correlation coefficients > 0.9 or <-0.9
plt.figure(figsize=(10,8))
plt.subplot(1,1,1)
sns.heatmap(corr[np.abs(corr) > 0.9])
plt.show()
```

```python
# As before, we create the correlation matrix
# and find rows and columns where correlation coefficients > 0.9 or <-0.9
corr = X_data.corr()
r, c = np.where(np.abs(corr) > 0.9)

# We are only interested in the off diagonal entries:
off_diagonal = np.where(r != c)

# Show the correlation matrix rows and columns where we have highly correlated off diagonal entries:
corr.iloc[r[off_diagonal], c[off_diagonal]]
```
